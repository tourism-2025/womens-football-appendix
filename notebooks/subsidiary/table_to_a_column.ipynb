```python
import pandas as pd
```

```python
# خواندن فایل اکسل
file_path = 'aaa.xlsx'  # مسیر فایل
df = pd.read_excel(file_path)

# تبدیل تمام ستون‌ها به یک ستون
stacked_data = pd.concat([df[col] for col in df.columns], ignore_index=True)

# ذخیره نتیجه در یک فایل جدید
stacked_data.to_excel('output.xlsx', index=False, header=False)

print("داده‌ها به صورت زیر هم ذخیره شدند.")
```
