```python
import pandas as pd

# خواندن فایل‌های اکسل
file1 = 'file1.xlsx'  # فایل اول (شامل ستون color)
file2 = 'file2.xlsx'  # فایل دوم (ستون color خالی)

# داده‌های فایل اول و دوم را بارگذاری کنید
df1 = pd.read_excel(file1)  # ستون‌های: name, number, color
df2 = pd.read_excel(file2)  # ستون‌های: Name_ID, Number_Value, color (خالی)

# ادغام داده‌ها بر اساس ستون‌های متفاوت
merged = pd.merge(
    df2,
    df1,
    left_on=['Name_ID', 'Number_Value'],  # نام ستون‌های فایل دوم
    right_on=['name', 'number'],          # نام ستون‌های فایل اول
    how='left'
)

# به‌روزرسانی ستون color فایل دوم
df2['color'] = merged['color']

# ذخیره فایل جدید
df2.to_excel('updated_file2.xlsx', index=False)

print("ستون color فایل دوم تکمیل شد و خروجی در 'updated_file2.xlsx' ذخیره شد.")
```
