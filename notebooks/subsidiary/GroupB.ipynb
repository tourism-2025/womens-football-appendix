```python
file_path = 'Specialized_questions.xlsx'
df = pd.read_excel(file_path)

# تبدیل به فرمت Long Format با مدیریت داده‌های خالی
long_format_data = []

for index, row in df.iterrows():
    respondent_id = row['Row']  # ستون شناسه پاسخ‌دهنده
    for col in df.columns[1:]:  # از ستون B1 به بعد
        ranks = str(row[col]).split(',')  # جداسازی بر اساس کاما
        for rank, option in enumerate(ranks, start=1):
            option = option.strip()  # حذف فاصله‌های اضافی
            if option.isdigit():  # بررسی اینکه مقدار عددی است یا نه
                long_format_data.append({
                    "شناسه": respondent_id,
                    "سوال": col,
                    "گزینه": int(option),
                    "رتبه": rank
                })

# تبدیل به دیتافریم
long_df = pd.DataFrame(long_format_data)

# ذخیره در فایل اکسل جدید
long_df.to_excel("long_format_dataset.xlsx", index=False)

print("دیتاست با موفقیت ایجاد شد!")
```
